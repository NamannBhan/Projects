```python
df=df[df['Volume']!=0]
df.reset_index(drop=True,inplace=True)
df.head(10)
```

```python
def create_label(df,lookahead=5,threshold=0.01):
  df['future_return']=df['Close'].pct_change(lookahead).shift(-lookahead)
  #assigning labels
  df['label']=0
  df.loc[df['future_return']>threshold,'label']=2
  df.loc[df['future_return']<-threshold,'label']=1
  df.drop(columns=['future_return'],inplace=True)
```

```python
create_label(df)
```

```python
df=df[df['label']!=0]
df
```

```python
import autobnn as ab
import jax
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
def create_sliding_window(df,window=50,lookahead=5,threshold=0.01):#window size means keeping 50 closing prices
  if 'Close' not in df.columns or 'label' not in df.columns:
    raise ValueError('DataFrame must contain "Close" and "Label" columns')

  create_label(df=df,lookahead=lookahead,threshold=threshold)

  closeprices=df['Close'].values
  labels=df['label'].values

  n_samples=len(df)-window
  if n_samples<=0:
    raise ValueError('not enough data to create sliding window')

  x_list=[]
  y_list=[]

  for i in range(n_samples):
    window_data=closeprices[i:i+window]
    last_candle_label=labels[i+window-1]#label of the last candle in window size
    x_list.append(window_data)
    y_list.append(last_candle_label)

  x=np.array(x_list)
  y=np.array(y_list)

  return x,y
```

```python
x,y=create_sliding_window(df)
len(x[0])
```

```python
from autobnn import estimators,operators,kernels
```

```python
def walk_forward_autobnn(df,window=100,train_size=50,step_size=5,threshold=0.01,n_classes=3):#classes coz labels are 1,2,3

  #step 1: building our dataset
  x,y=create_sliding_window(df=df,window=window,threshold=threshold)
  n_total=len(x)
  if n_total<train_size+1:
    raise ValueError('Not enough samples for walk-forward analysis')

  accuracies=[]

  #step 2: walk-forward loop
  for i in range(0,n_total-train_size,step_size):
    x_train=x[i:i+train_size]
    y_train=y[i:i+train_size]

    test_index=i+train_size
    test_x=x[test_index:test_index+1]
    test_y=y[test_index:test_index+1]

    #scaling
    scaler=StandardScaler()
    train_x_scaled=scaler.fit_transform(x_train)
    test_x_scaled=scaler.transform(test_x)


    # step 3: training one model per class
    estimators=[]
    for c in range(n_classes):
      y_train_c=(y_train==c).astype(int) #multi-class into binary classification one vs rest

      model_c=ab.operators.Add(
          bnns=(
                ab.kernels.PeriodicBNN(width=20,period=12),#width=20 for smoothing
                ab.kernels.LinearBNN(width=20),
                ab.kernels.MaternBNN(width=20)
          ))
      estimator_c=ab.estimators.AutoBnnMapEstimator(model_c,
                                                    likelihood_model='normal_likelihood_logistic_noise',
                                                    seed=jax.random.PRNGKey(42),
                                                    periods=[12])
      estimator_c.fit(train_x_scaled,y_train_c)
      estimators.append(estimator_c)


    #step 4: predict probabilites for each class on single test sample
    class_prob=[]
    for c in range(n_classes):
      y_pred_c=estimators[c].predict_proba(test_x_scaled)

      if y_pred_c.ndim==2:
        logit=y_pred_c[0,0]
      else:
        logit=y_pred_c[0]

      prob_c=1.0/(1.0+np.exp(-logit))
      class_prob.append(prob_c)

    #step 5: picking class with highest prob
    y_pred_class=np.argmax(class_prob)

    #step 6: single sample acc
    acc=accuracy_score(test_y,[y_pred_class])
    accuracies.append(acc)

  return accuracies
```

```python
dfsample=df[:1000].copy()

results=walk_forward_autobnn(df=dfsample,window=40,train_size=20,step_size=5,threshold=0.01,n_classes=3)
print('Accuracies',results)
if results:
  print('Mean Accuracy',np.mean(results))
```
